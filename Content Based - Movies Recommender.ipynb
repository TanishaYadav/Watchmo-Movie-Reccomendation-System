# Movies Recommendation System 

```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
movies.head()
```

```python
credits.head()
```

## Merging datsets

```python
movies.rename(columns = {'id': 'movie_id'}, inplace = True)
movies = movies.merge(credits, on = 'movie_id')
movies.head()
```

## Considering useful features

```python
df = movies[['movie_id','title_x', 'overview', 'genres','keywords','cast','crew']]
df.head()  # columns that will be helpful in determinig tags
```

## Missing Values

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

## Extracting genres 

```python
eval(df['genres'][0])
```

```python
def get_tags(movie):
    genres = []
    for i in eval(movie):
        genres.append(i['name'])
    return genres
        
```

```python
get_tags(df['genres'][0])
```

```python
df['genres'] = df['genres'].apply(get_tags)
df.head(1)
```

```python
df['keywords'] = df['keywords'].apply(get_tags)
df.head(1)
```

```python
eval(df['cast'][0])[:3]  # information of top 3 actors
```

```python
def get_actors_name(movie):
    actors = []
    for i in eval(movie)[:3]:
        actors.append(i['name'])
    return actors   
```

```python
get_actors_name(df['cast'][0])
```

```python
df['cast'] = df['cast'].apply(get_actors_name)
```

```python
df.head(1)
```

```python
eval(df['crew'][0])
```

```python
def get_director(crew):
    directors = []
    for i in eval(crew):
        if i['job'] == 'Director':
            directors.append(i['name'])
            break
    return directors
```

```python
get_director(df['crew'][0])
```

```python
df['crew'] = df['crew'].apply(get_director)
```

```python
df.head(1)
```

```python
df['overview'][0].split(' ')
```

```python
df['overview'] = df['overview'].apply(lambda x: x.split(' '))
```

```python
df.head(1)
```

```python
df['cast'][0][0].replace(' ','')
```

```python
df['cast'] =   df['cast'].apply(lambda x: [i.replace(" ",'') for i in x])
df['crew'] =   df['crew'].apply(lambda x: [i.replace(" ",'') for i in x])
df['genres']  = df['genres'].apply(lambda x: [i.replace(" ",'') for i in x])
df['keywords'] =   df['keywords'].apply(lambda x: [i.replace(" ",'') for i in x])
```

```python
df.head(1)
```

```python
['fsfd','fsdf']+['erw','as']
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
df.head(1)
```

```python
data = df[['movie_id','title_x','tags']]
data.head(1)
```

```python
data['tags'] = data['tags'].apply(lambda x: " ".join(x))
data.head(1)
```

```python
data['tags']
```

```python
data['tags'] = data['tags'].apply(lambda x:x.lower())
data.head(1)
```

```python
for i in data['tags'][:3]:
    print((i.split()))
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
    stemmed = []
    for i in text.split(" "):
        stemmed.append(ps.stem(i))
    return " ".join(stemmed)
```

```python
data['tags'] = data['tags'].apply(stemming)
data.head(1)
```

## Bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
vectors = cv.fit_transform(data['tags']).toarray()
vectors.shape
```

```python
vectors  # a vector for each movie tag
```

For higher dimensions, Euclidian distance is not a good choice. Therefore, here we are using Cosine Similarity (angle between vectors).

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity[data[data['title_x'] == 'Gandhi'].index[0]]
```

```python
sorted(list(enumerate(similarity[data[data['title_x'] == 'Gandhi'].index[0]])), reverse = True, key = lambda x : x[1])[1:6]
```

```python
for i in sorted(list(enumerate(similarity[data[data['title_x'] == 'Gandhi'].index[0]])), reverse = True, key = lambda x : x[1])[1:6]:
    print(data.iloc[i[0]]['title_x'])
```

```python
def recommend(movie):
    recommeandations = []
    movie_index = data[data['title_x'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x : x[1])[1:6]
    
    for i in movies_list:
        recommeandations.append(data.iloc[i[0]]['title_x'])
    return recommeandations

recommend('Gandhi')
        
```
